```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn import decomposition
from random import sample
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn.neural_network import MLPClassifier
from sklearn import preprocessing
```

```python
datos = pd.read_excel('wine.xlsx')
datos
```

```python
# DistribuciÃ³n
sns.countplot(datos['quality'])
```

```python
#BOXPLOT de cada caracteristica con la variable diagnosis
for i in ['fixed acidity','volatile acidity','citric acid','residual sugar','chlorides','free sulfur dioxide','total sulfur dioxide','density','pH','sulphates','alcohol']:
    boxplot = datos.boxplot(column=i, by='quality', fontsize=10,figsize=(10,5),)
    plt.show()
```

```python
# CorrelaciÃ³n
correlacion_estudiantes_base=datos.corr()
correlacion_estudiantes_base.style.background_gradient(cmap='coolwarm')
```

```python
data=datos.drop(['volatile acidity','density','alcohol'],axis=1)
data 
```

```python
x = data.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
df = pd.DataFrame(x_scaled)
df.columns = ['fixed aciditu', 'citric acid',' residual sugar','chlorides','free sulfur dioxide','total sulfur dioxide', 'pH', 'sulphates', 'quality']
datos = df.drop(['quality'], axis=1)
columna_quality = pd.DataFrame(data['quality'])
dato_normalizado = pd.concat([datos,columna_quality], axis=1)
dato_normalizado.head()
```

```python
X=dato_normalizado.iloc[:,0:8] ## Corresponde a las caracteristicas sin la variable de llegada diagnosis
X
```

```python
Y = dato_normalizado.iloc[:,8:9]
Y
```

```python
# PCA 
pca = decomposition.PCA(n_components=8)
X1 = pca.fit_transform(X)
```

```python
X_FSEX =pd.DataFrame({'PCA1': X1[:,0], 'PCA2': X1[:,1],'PCA3': X1[:,2], 'PCA4': X1[:,3],'PCA5': X1[:,4],'PCA6': X1[:,5],'PCA7': X1[:,6],'PCA8': X1[:,7]})
X_FSEX
```

```python
X_con_F = pd.concat([X_FSEX, X], axis=1)
X_con_F
```

```python
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
from sklearn.svm import SVC
# Sequential Forward Selection(sfs)
sfs = SFS(MLPClassifier(activation='logistic',solver='lbfgs',alpha=1e-05,hidden_layer_sizes=(30,)),
           k_features=8,
           forward=True,
           floating=False,
           scoring = 'r2',
           cv = 0)
```

```python
# Sequential Forward Selection(sfs)
sbs = SFS(MLPClassifier(activation='logistic',solver='lbfgs',alpha=1e-05,hidden_layer_sizes=(30,)),
           k_features=8,
           forward=False,
           floating=False,
           scoring = 'r2',
           cv = 0)
```

```python
#FSEs-sfs
sfs.fit (X, Y) 
sfs.k_feature_names_ 
```

```python
#FSEs-sbs
sbs.fit (X, Y) 
sbs.k_feature_names_ 
```

```python
#FSXs-sfs
sfs.fit (X_FSEX, Y) 
sfs.k_feature_names_ 
```

```python
#FSXs-sbs
sbs.fit (X_FSEX, Y) 
sbs.k_feature_names_
```

```python
#FSXs:FSEs-sbs
sbs.fit (X_con_F, Y) 
sbs.k_feature_names_
```

```python
#FSXs:FSEs-sfs
sfs.fit (X_con_F, Y) 
sfs.k_feature_names_
```

```python
import copy
Xsfs = copy.copy(X_con_F)
Xsbs = copy.copy(X_con_F)
Xsbs = Xsbs.drop(columns=['PCA1','PCA4','PCA6','PCA7','pH',' residual sugar','chlorides','free sulfur dioxide'], axis=1)
Xsfs = Xsfs.drop(columns=['PCA1','PCA2','PCA3','PCA8','free sulfur dioxide','sulphates','total sulfur dioxide',' residual sugar'], axis=1)
# 'radius_mean','texture_mean','perimeter_mean','area_mean','smoothness_mean','compactness_mean','concavity_mean','concave_points_mean','symmetry_mean','radius_se','perimeter_se','area_se','compactness_se','concavity_se','concave_points_se','radius_worst','texture_worst','perimeter_worst','area_worst','smoothness_worst','compactness_worst','concavity_worst','concave_points_worst','symmetry_worst','fractal_dimension_worst'
```

```python
X_test,X_train,Y_test,Y_train=train_test_split(X,Y,test_size=0.2,random_state=40) # fse
X_FSEX_test,X_FSEX_train,Ye_test,Ye_train=train_test_split(X_FSEX,Y,test_size=0.2,random_state=40) #fsx
X_con_F_test,X_con_F_train,yef_test,yef_train=train_test_split(X_con_F,Y,test_size=0.2,random_state=40) #fsx:fse
Xsbs_test,Xsbs_train,ysbs_test,ysbs_train=train_test_split(Xsbs,Y,test_size=0.2,random_state=40) # (fse:fsx))s(sbs)
Xsfs_test,Xsfs_train,ysfs_test,ysfs_train=train_test_split(Xsfs,Y,test_size=0.2,random_state=40) # (fse:fsx)s(sfs)
```

```python
cls1=MLPClassifier(activation='logistic',solver='lbfgs',alpha=1e-05,hidden_layer_sizes=(30,))
cls2=MLPClassifier(activation='logistic',solver='lbfgs',alpha=1e-05,hidden_layer_sizes=(30,))
cls3=MLPClassifier(activation='logistic',solver='lbfgs',alpha=1e-05,hidden_layer_sizes=(30,))
cls4=MLPClassifier(activation='logistic',solver='lbfgs',alpha=1e-05,hidden_layer_sizes=(30,))
cls5=MLPClassifier(activation='logistic',solver='lbfgs',alpha=1e-05,hidden_layer_sizes=(30,))
        
```

```python
cls1.fit(X_train,Y_train)
cls3.fit(X_FSEX_train.abs(),Ye_train)
cls4.fit(X_con_F_train.abs(),yef_train)
cls4.fit(Xsbs_train.abs(),ysbs_train)
cls5.fit(Xsfs_train.abs(),ysfs_train)
```

```python
from sklearn import metrics
predict_test1 = cls1.predict(X_test)
predict_test2 = cls2.predict(X_FSEX_test)
predict_test3 = cls3.predict(X_con_F_test)
predict_test4 = cls4.predict(Xsbs_test)
predict_test5 = cls5.predict(Xsfs_test)
#Evaluar
acc_n1 = metrics.accuracy_score(predict_test1,Y_test)
acc_n2 = metrics.accuracy_score(predict_test2,Ye_test)
acc_n3 = metrics.accuracy_score(predict_test3,yef_test)
acc_n4 = metrics.accuracy_score(predict_test4,ysbs_train)
acc_n5 = metrics.accuracy_score(predict_test5,ysfs_train)
```

```python
print("Accuracy sin normalizaciÃ³n: {:.1f}".format(acc_n1*100))
```
